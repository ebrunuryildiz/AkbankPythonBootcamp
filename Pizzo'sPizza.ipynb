```python
import csv
from datetime import datetime       

class bcolors:
    HEADER = '\033[95m'
    OKBLUE = '\033[94m'
    OKCYAN = '\033[96m'
    OKGREEN = '\033[92m'
    WARNING = '\033[93m'
    FAIL = '\033[91m'
    ENDC = '\033[0m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'

class Pizza:
    def __init__(self, description):
        self.description = description
        self.cost = 0
        
    def get_description(self):
        return self.description
    
    def get_cost(self):
        return self.cost

class ClassicPizza(Pizza):
    def get_cost(self):
        return 10.00

class MargheritaPizza(Pizza):
    def get_cost(self):
        return 20.00        
    
class TurkPizza(Pizza):
    def get_cost(self):
        return 15.00            

class DominosPizza(Pizza):
    def get_cost(self):
        return 40.00    
    
classic_pizza    = ClassicPizza("The classsic cheese and tomato base, topped with basil leaves")    
margherita_pizza = MargheritaPizza("Fresh mozzarella, tomato sauce, fresh basil, olive oil & provolone") 
turk_pizza       = TurkPizza("Couscous, lamp mince, harissa, tomato paste, red onion")
dominos_pizza    = DominosPizza("Tomato sauce, basil leaves, oregano, mozzarella, pepperoni, chichken, tomato, olive")    

class Sauce(Pizza):
    def __init__(self, component):
        self.component = component

    def get_cost(self):
        return self.component.get_cost() + Pizza.get_cost(self)

    def get_description(self):
        return self.component.get_description() + ' ' + Pizza.get_description(self)

class Olives(Sauce):
    def __init__(self, component):
        super().__init__(component)
        self.description = "Olives"
        self.cost = 1.50

    def get_cost(self):
        return self.component.get_cost() + Pizza.get_cost(self) + self.cost

    def get_description(self):
        return self.component.get_description() + ' ' + Pizza.get_description(self) + ', ' + self.description

class Mushrooms(Sauce):
    def __init__(self, component):
        super().__init__(component)
        self.description = "Mushrooms"
        self.cost = 1.00

    def get_cost(self):
        return self.component.get_cost() + Pizza.get_cost(self) + self.cost

    def get_description(self):
        return self.component.get_description() + ' ' + Pizza.get_description(self) + ', ' + self.description

class GoatCheese(Sauce):
    def __init__(self, component):
        super().__init__(component)
        self.description = "Goat Cheese"
        self.cost = 2.00

    def get_cost(self):
        return self.component.get_cost() + Pizza.get_cost(self) + self.cost

    def get_description(self):
        return self.component.get_description() + ' ' + Pizza.get_description(self) + ', ' + self.description

class Meat(Sauce):
    def __init__(self, component):
        super().__init__(component)
        self.description = "Meat"
        self.cost = 3.50

    def get_cost(self):
        return self.component.get_cost() + Pizza.get_cost(self) + self.cost

    def get_description(self):
        return self.component.get_description() + ' ' + Pizza.get_description(self) + ', ' + self.description

class Onions(Sauce):
    def __init__(self, component):
        super().__init__(component)
        self.description = "Onions"
        self.cost = 0.50

    def get_cost(self):
        return self.component.get_cost() + Pizza.get_cost(self) + self.cost

    def get_description(self):
        return self.component.get_description() + ' ' + Pizza.get_description(self) + ', ' + self.description

class Corn(Sauce):
    def __init__(self, component):
        super().__init__(component)
        self.description = "Corn"
        self.cost = 1.00

    def get_cost(self):
        return self.component.get_cost() + Pizza.get_cost(self) + self.cost

    def get_description(self):
        return self.component.get_description() + ' ' + Pizza.get_description(self) + ', ' + self.description

def read_menu(filename):
    with open(filename, 'r') as f:
        menu = f.read()
    return menu

def print_menu():
    menu = read_menu('Menu.txt')
    print(menu)
    ##print(bcolors.OKBLUE + menu)

def get_customer_info():
    name = input("Name: ")
    id_number = input("ID number: ")
    credit_card_number = input("Credit card number: ")
    credit_card_password = input("Credit card password: ")
    return name, id_number, credit_card_number, credit_card_password

def save_order_to_database(name, id_number, credit_card_number, credit_card_password, pizza_description):
    with open('Orders_Database.csv', 'a', newline='') as csvfile:
        fieldnames = ['Name', 'ID Number', 'Credit Card Number', 'Order Description', 'Time Ordered', 'Credit Card Password']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        if csvfile.tell() == 0:
            writer.writeheader()
        writer.writerow({
            'Name': name,
            'ID Number': id_number,
            'Credit Card Number': credit_card_number,
            'Order Description': pizza_description,
            'Time Ordered': datetime.now().strftime('%Y-%m-%d %H:%M:%S'),
            'Credit Card Password': credit_card_password           
        })

def print_welcomeMessages():
    print(bcolors.HEADER + "           -            ")
    print(bcolors.HEADER + "         -   -          ")
    print(bcolors.HEADER + "       -   .   -        ")
    print(bcolors.HEADER + "     -   . ! .   -   Pizzo's Pizza   ")
    print(bcolors.HEADER + "       -   .   -        Since 1864   ")
    print(bcolors.HEADER + "         -   -          ")
    print(bcolors.HEADER + "           -            ")    

def main():
    print_welcomeMessages()
    print_menu()

    # Take user input for pizza and sauce selection
    while True:
        try:
            pizza_choice = int(input(bcolors.OKBLUE + "Please choose a pizza (1-4): "))
            if pizza_choice not in range(1, 5):
                raise ValueError()
            break
        except ValueError:
            print("Invalid input! Please choose a pizza between 1 and 4.")
            5

    while True:
        try:
            sauce_choice = int(input("Please choose a sauce (11-16): "))
            if sauce_choice not in range(11, 17):
                raise ValueError()
            break
        except ValueError:
            print("Invalid input! Please choose a sauce between 11 and 16.")

    # Calculate total price
    selectedPizza = None
    if pizza_choice == 1:
        selectedPizza = classic_pizza        
    elif pizza_choice == 2:
        selectedPizza = margherita_pizza
    elif pizza_choice == 3:
        selectedPizza = turk_pizza
    elif pizza_choice == 4:
        selectedPizza = dominos_pizza

    if sauce_choice == 11:
        selectedPizza = Olives(selectedPizza)
        print(selectedPizza.get_description())
        print(selectedPizza.get_cost())
    elif sauce_choice == 12:
        selectedPizza = Mushrooms(selectedPizza)
    elif sauce_choice == 13:
        selectedPizza = GoatCheese(selectedPizza)
    elif sauce_choice == 14:
        selectedPizza = Meat(selectedPizza)
    elif sauce_choice == 15:
        selectedPizza = Onions(selectedPizza)
    elif sauce_choice == 16:
        selectedPizza = Corn(selectedPizza)
    
    total_price = selectedPizza.get_cost()
    pizza_description = selectedPizza.get_description()
    print(bcolors.WARNING + f"Total price: {total_price:.2f}")
    
    name, id_number, credit_card_number, credit_card_password = get_customer_info()
    
    save_order_to_database(name, id_number, credit_card_number, credit_card_password, pizza_description)
    print(bcolors.OKGREEN + "Your order has been saved to the database.")

if __name__ == '__main__':
    main()
```
